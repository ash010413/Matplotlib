```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('title')
ax.set_ylabel('y')
ax.set_xlabel('x')
```

```python
fig = plt.figure()
ax1= fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
ax1.plot(x,y)
ax2.plot(x,y)
fig
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
ax1.plot(x,z)
ax1.set_ylim([0,10000])
ax1.set_xlim([0,100])
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
fig
```

```python
plt.subplots(nrows=1,ncols=2)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,lw=5,ls='--')

plt.subplot(1,2,2)
plt.plot(x,z,lw=2,ls='-')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,4))
axes[0].plot(x,y)
axes[1].plot(x,z)
```
